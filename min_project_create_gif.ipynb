```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
scooby = np.array(plt.imread("dog.jpeg"))
plt.imshow(scooby)
```

```python
image_rotate = np.transpose(scooby,(1,0,2))
plt.imshow(image_rotate)
```

```python
plt.imshow(scooby[::-1,:,:])
```

```python
image_right_rotate = np.transpose(scooby,(1,0,2))[:,::-1,:]
plt.imshow(image_right_rotate)
```

```python
frame = []
# plt.imshow(scooby)
frame.append(Image.fromarray(scooby))

image_rotate = np.transpose(scooby,(1,0,2))
# plt.imshow(image_rotate)
frame.append(Image.fromarray(image_rotate))

# plt.imshow(scooby[::-1,:,:])
frame.append(Image.fromarray(scooby[::-1,:,:]))

image_right_rotate = np.transpose(scooby,(1,0,2))[:,::-1,:]
# plt.imshow(image_right_rotate)
frame.append(Image.fromarray(image_right_rotate))
```

```python
frame
```

```python
frame[0].save('scooby_rotation.gif', save_all=True, append_images=frame[1:], duration=500, loop=0)
```
